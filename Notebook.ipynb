```python
import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('data_set.csv')
```

```python
df.head()
```

```python
df.drop(columns=['Timestamp'], inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={'What is your gender?': 'Gender'}, inplace=True)
df.rename(columns={'How many classes have you attended out of the total number of classes held for the Data Structures course?': 'Attendance'}, inplace=True)
df.rename(columns={'What was your score on the final exam for the Data Structures course?': 'Final_Score'}, inplace=True)
df.rename(columns={'What was your overall score on the coursework assignments for the Data Structures course?': 'Coursework_Score'}, inplace=True)
df.rename(columns={'On the night before the final exam, how many hours did you spend studying for the Data Structures exam?': 'Studing_Hours'}, inplace=True)
df.rename(columns={'On the night before the final exam, how many hours did you sleep?': 'Sleeping_Hours'}, inplace=True)
```

# Measures of Central Tendency
### Final Exam Score

```python
# Mean
mean_score = df['Final_Score'].mean()

# Median
median_score = df['Final_Score'].median()

# Mode
mode_score = stats.mode(df['Final_Score'])

# Standard Deviation
std_score = df['Final_Score'].std()

print("Mean:", mean_score)
print("Median:", median_score)
print("Mode:", mode_score.mode)
print("Standard Deviation:", std_score)
```

### Coursework Score

```python
# Mean
mean_score = df['Coursework_Score'].mean()

# Median
median_score = df['Coursework_Score'].median()

# Mode
mode_score = stats.mode(df['Coursework_Score'])

# Standard Deviation
std_score = df['Coursework_Score'].std()

print("Mean:", mean_score)
print("Median:", median_score)
print("Mode:", mode_score.mode)
print("Standard Deviation:", std_score)
```
